```python
import requests
import time
import urllib.request
import json

url = "http://api.open-notify.org/iss-now.json"
data_count = 100 #used chat gpt for this
interval = 17 #used chat gpt for this, will wait approx 28 minutes to get 100 data

data = []
for _ in range(data_count):
    response = requests.get(url)
    if response.status_code == 200:
        obj = json.loads(response.content.decode())
        data_point = {
            "timestamp": obj['timestamp'],
            "latitude": obj['iss_position']['latitude'],
            "longitude": obj['iss_position']['longitude']
        }
        data.append(data_point)
    time.sleep(interval)

for data_point in data:
    print("timestamp:", data_point['timestamp'])
    print("latitude:", data_point['latitude'], ", longitude:", data_point['longitude'])
    print()  # Add an empty line between data points
```

```python
import matplotlib.pyplot as plt


timestamps = [data_point['timestamp'] for data_point in data]
latitudes = [data_point['latitude'] for data_point in data]
longitudes = [data_point['longitude'] for data_point in data]


plt.plot(longitudes, latitudes)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Location of the International Space Station')
plt.grid(True)
plt.xticks(range(-200, 200, 50))
plt.yticks(range(10, 100, 10))
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()


for i, txt in enumerate(timestamps): #chatgpt this for the timestamp
    plt.annotate(txt, (longitudes[i], latitudes[i]), fontsize=8)

plt.show()
```
